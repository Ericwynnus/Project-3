```python
df = pd.read_csv("steam-200k.csv")
df.isnull().sum()
```

```python
df.rename(columns= {"151603712":"user id","The Elder Scrolls V Skyrim":"game title","purchase":"purchase","1":"hours","0":"value"}, inplace = True)
df.head()
```

```python
#Players Can be assigned ID's we are going to suggest the games based on the number 
# of hours they play and what sort of game.

df.describe()
```

```python
# Players can be filtered by volume of hours played/engaged per game. Here 10 hours or more is used.

df_final = df.query("hours>10")
#df_final = df.query("behavior_name>10")
```

```python
df_final.rename(columns = {"1":"time"})
df_final.drop("value",axis=1,inplace= True)
```

```python
df_final.drop("purchase",axis = 1, inplace =True)
```

```python
df_final.head()
```

```python
# now we will build the recommendation system for the above data using KNN Algorithm

# The KNN Algorithim (K-Nearest Neighbor) is used for classification and looks to assigns new data based on
# on how close or similiar the data is to the points in the training data. 

pivot_knn = df_final.pivot_table(index= "game title", columns = "user id", values="hours").fillna(0)
pivot_knn.head()
```

```python
#The Brute Force Algorithim is an example of a Optimal Algorithim because it is designed to produce an optimal result.

#Other possible algorithims examples would be BallTree or KDTree.

from scipy.sparse import csr_matrix
csr_knn = csr_matrix(pivot_knn)
from sklearn.neighbors import NearestNeighbors
nbrs = NearestNeighbors(metric = "cosine", algorithm = "brute")
nbrs.fit(csr_knn)
```

```python
pivot_corr = df_final.pivot_table(index= "user id",columns="game title",values = "hours").fillna(0)
pivot_corr.head()
```

```python
# In this query the game "7 Days to Die" has emerged as a popular title.
# This game is now used to search for similiar titles.

corr_mat = pivot_corr["7 Days to Die"]
corr_out = pivot_corr.corrwith(corr_mat)
game_corr_df = pd.DataFrame(corr_out,columns= ["corelation"])
rec_corr = game_corr_df.sort_values(ascending = False,by= "corelation").head(5).index
```

```python
distances,indices = nbrs.kneighbors(pivot_knn.iloc[7].values.reshape(1,-1))
rec_1 = []
for i in indices:
    x = pivot_knn.index[i].values
    rec_1.append(x)
rec_1 = list(rec_1)    
rec_knn = []
for i in range(5):
    rec_knn.append(rec_1[0][i])
```

```python
for i in rec_knn:
    if i in rec_corr:
        print(i)
```
